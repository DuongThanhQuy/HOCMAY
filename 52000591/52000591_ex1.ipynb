```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm số cần tối ưu hóa: f(x) = x^2
def function(x):
    return x**2

# Đạo hàm của hàm số
def gradient(x):
    return 2 * x

# Thuật toán Gradient Descent
def gradient_descent(learning_rate, epochs):
    x = 10  # Giá trị khởi tạo của x
    x_values = []
    y_values = []

    for _ in range(epochs):
        x -= learning_rate * gradient(x)
        y = function(x)
        x_values.append(x)
        y_values.append(y)

    return x_values, y_values

# Thiết lập tham số cho Gradient Descent
learning_rate = 0.1
epochs = 50

# Chạy thuật toán Gradient Descent
x_values, y_values = gradient_descent(learning_rate, epochs)

# Vẽ biểu đồ minh họa
plt.figure(figsize=(8, 6))
plt.title('Gradient Descent Optimization')
plt.xlabel('Epochs')
plt.ylabel('f(x)')
plt.scatter(x_values, y_values, c=range(len(x_values)), cmap='viridis', label='Iterations')
plt.colorbar(label='Iterations')
plt.plot(x_values, y_values, linestyle='dashed', color='black', alpha=0.5, label='Optimization Path')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm số cần tối ưu
def function_to_optimize(x):
    return x ** 2

# Đạo hàm của hàm số cần tối ưu
def derivative_function(x):
    return 2 * x

# Gradient Descent (GD)
def gradient_descent(learning_rate, epochs):
    x_gd = 5  # Khởi tạo giá trị ban đầu
    x_values_gd = []
    for epoch in range(epochs):
        gradient = derivative_function(x_gd)
        x_gd = x_gd - learning_rate * gradient
        x_values_gd.append(x_gd)
    return x_values_gd

# Stochastic Gradient Descent (SGD)
def stochastic_gradient_descent(learning_rate, epochs):
    x_sgd = 5  # Khởi tạo giá trị ban đầu
    x_values_sgd = []
    for epoch in range(epochs):
        for _ in range(100):  # Sử dụng 100 điểm dữ liệu trong mỗi epoch
            random_point = np.random.uniform(-10, 10)  # Tạo ngẫu nhiên điểm dữ liệu
            gradient = derivative_function(random_point)
            x_sgd = x_sgd - learning_rate * gradient
            x_values_sgd.append(x_sgd)
    return x_values_sgd

# Cài đặt các tham số
learning_rate = 0.1
epochs = 100

# Chạy thuật toán và lấy các giá trị của x trong quá trình học
x_values_gd = gradient_descent(learning_rate, epochs)
x_values_sgd = stochastic_gradient_descent(learning_rate, epochs)

# Vẽ biểu đồ minh họa
plt.figure(figsize=(10, 6))
plt.scatter(range(len(x_values_gd)), x_values_gd, color='blue', label='Gradient Descent')
plt.scatter(range(len(x_values_sgd)), x_values_sgd, color='red', label='Stochastic Gradient Descent')
plt.title('Optimization using Gradient Descent vs Stochastic Gradient Descent')
plt.xlabel('Iterations')
plt.ylabel('Value of x')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm mục tiêu: f(x) = x^2
def objective_function(x):
    return x**2

# Đạo hàm của hàm mục tiêu: f'(x) = 2x
def gradient(x):
    return 2 * x

# Thuật toán Gradient Descent với Momentum
def gradient_descent_momentum(learning_rate, gamma, iterations):
    x = 10  # Khởi tạo x ngẫu nhiên
    velocity = 0

    x_history = []
    for i in range(iterations):
        grad = gradient(x)
        velocity = gamma * velocity + learning_rate * grad
        x = x - velocity
        x_history.append(x)

    return x_history

# Tham số của thuật toán
learning_rate = 0.1
gamma = 0.9
iterations = 50

# Tối ưu hóa và lưu lịch sử tọa độ
history = gradient_descent_momentum(learning_rate, gamma, iterations)

# Vẽ biểu đồ
plt.figure(figsize=(8, 6))
plt.title('Tối ưu hóa bằng Gradient Descent với Momentum')
plt.xlabel('Iterations')
plt.ylabel('Giá trị của x')
plt.plot(history, 'o-', color='blue', label='Đà Momentum')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm số cần tối ưu
def function_to_optimize(x):
    return x**2

# Gradient của hàm số
def gradient(x):
    return 2 * x

# Thuật toán Adagrad
def adagrad(learning_rate, epsilon, iterations):
    x = 10  # Giá trị khởi tạo của x
    G = 0   # Khởi tạo ma trận chéo G
    x_values = []
    for i in range(iterations):
        grad = gradient(x)
        G += grad**2
        x -= (learning_rate / np.sqrt(G + epsilon)) * grad
        x_values.append(x)
    return x_values

# Thực thi thuật toán Adagrad
learning_rate = 0.1
epsilon = 1e-8
iterations = 100

x_values = adagrad(learning_rate, epsilon, iterations)

# Vẽ biểu đồ
plt.figure(figsize=(8, 6))
plt.scatter(range(iterations), x_values, c=range(iterations), cmap='viridis')
plt.colorbar(label='Iterations')
plt.xlabel('Iterations')
plt.ylabel('Value of x')
plt.title('Convergence using Adagrad')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm số cần tối ưu hóa: f(x) = x^2
def function(x):
    return x**2

# Gradient của hàm số
def gradient(x):
    return 2 * x

# Thuật toán RMSprop
def rmsprop(initial_x, learning_rate, epochs):
    x = initial_x
    epsilon = 1e-8
    decay_rate = 0.9
    cache = 0
    
    x_values = []
    y_values = []

    for i in range(epochs):
        grad = gradient(x)
        cache = decay_rate * cache + (1 - decay_rate) * grad**2
        x -= (learning_rate / np.sqrt(cache + epsilon)) * grad
        
        # Lưu lại các giá trị của x và f(x)
        x_values.append(x)
        y_values.append(function(x))
    
    return x_values, y_values

initial_x = 4
learning_rate = 0.1
epochs = 50

x_values, y_values = rmsprop(initial_x, learning_rate, epochs)

plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values, c=np.arange(len(x_values)), cmap='viridis')
plt.plot(x_values, y_values, linestyle='-', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Tối ưu hóa hàm số f(x) = x^2 bằng RMSprop')
plt.colorbar(label='Epochs')
plt.grid(True)
plt.show()
```
